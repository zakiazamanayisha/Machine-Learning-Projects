```python
occup=pd.get_dummies(data.occupation,dtype=int).add_prefix('occupation_')
```

```python
data=pd.concat([data.drop('occupation',axis=1),occup],axis=1)
```

```python
data
```

```python
workcl=pd.get_dummies(data['workclass'],dtype=int).add_prefix('workclass_')
marital_stat=pd.get_dummies(data['marital-status'],dtype=int).add_prefix('marital-status')
rltn=pd.get_dummies(data['relationship'],dtype=int).add_prefix('relationship_')
race=pd.get_dummies(data['race'],dtype=int).add_prefix('race_')
nat_coun=pd.get_dummies(data['native-country'],dtype=int).add_prefix('native-country_')
```

```python
data=pd.concat([data.drop('workclass',axis=1),workcl],axis=1)
data=pd.concat([data.drop('marital-status',axis=1),marital_stat],axis=1)
data=pd.concat([data.drop('relationship',axis=1),rltn],axis=1)
data=pd.concat([data.drop('race',axis=1),race],axis=1)
data=pd.concat([data.drop('native-country',axis=1),nat_coun],axis=1)
data=data.drop('education',axis=1)
```

```python
data.head()
```

```python
data['gender']=data['gender'].apply(lambda x: 1 if x=='Male' else 0)
data['income']=data['income'].apply(lambda x: 1 if x=='>50K' else 0)
```

```python
data.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

data=data.drop('fnlwgt',axis=1)
train_data,test_data=train_test_split(data,test_size=0.2)
```

```python
train_data
```

```python
test_data
```

```python
train_x=train_data.drop('income',axis=1)
train_y=train_data['income']
test_x=test_data.drop('income',axis=1)
test_y=test_data['income']
```

```python
forest=RandomForestClassifier()

forest.fit(train_x,train_y)
```

```python
forest.score(test_x,test_y)
```

```python
forest.feature_importances_
```

```python
forest.feature_names_in_
```

```python
importances=dict(zip(forest.feature_names_in_,forest.feature_importances_))
```

```python
importances={k:v for k,v in sorted(importances.items(),key=lambda x:x[1], reverse=True)}
```

```python
importances
```

```python
from sklearn.model_selection import GridSearchCV

param_grid={
    'n_estimators':[50,100,250],
    'max_depth':[5,10,20,None],
    'min_samples_split':[2,4],
    'max_features':['sqrt','log2']
}

grid_search=GridSearchCV(estimator=RandomForestClassifier(),
                       param_grid=param_grid,verbose=10)
```

```python
grid_search.fit(train_x,train_y)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_estimator_.score(test_x,test_y)
```
