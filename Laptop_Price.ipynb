```python
def turn_memory_into_MB(value):
    if "GB" in value:
        return float(value[:value.find("GB")])*1000
    elif "TB" in value:
        return float(value[:value.find("TB")])*1000000
```

```python
price['MemorySize']=price['MemorySize'].apply(turn_memory_into_MB)
```

```python
selected_features=target_corr[-27:].index
```

```python
limited_price=price[selected_features]
```

```python
limited_price
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(limited_price.corr(),annot=True,cmap='viridis')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

x,y=limited_price.drop('Price_euros',axis=1), limited_price['Price_euros']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15)

scaler=StandardScaler()

x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
forest=RandomForestRegressor()

forest.fit(x_train_scaled,y_train)
```

```python
forest.score(x_test_scaled,y_test)
```

```python
y_pred=forest.predict(x_test_scaled)
```

```python
plt.scatter(y_pred,y_test)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid={
    "n_estimators":[30,50,100],
    "max_features":[8,12,24],
}
grid_search=GridSearchCV(forest,param_grid,cv=5,scoring='neg_mean_squared_error',return_train_score=True)

grid_search.fit(x_train_scaled,y_train)
```

```python
grid_search.best_estimator_.score(x_test_scaled,y_test)
```

```python
y_pred2=grid_search.predict(x_test_scaled)
```

```python
plt.scatter(y_pred2,y_test)
```
