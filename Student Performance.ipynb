```python
data['school']=data['school'].apply(lambda x: 1 if x=='GP' else 0)
data['sex']=data['sex'].apply(lambda x: 1 if x=='M' else 0)
data['address']=data['address'].apply(lambda x: 1 if x=='U' else 0)
data['famsize']=data['famsize'].apply(lambda x: 1 if x=='GT3' else 0)
data['Pstatus']=data['Pstatus'].apply(lambda x: 1 if x=='T' else 0)
data['schoolsup']=data['schoolsup'].apply(lambda x: 1 if x=='yes' else 0)
data['famsup']=data['famsup'].apply(lambda x: 1 if x=='yes' else 0)
data['paid']=data['paid'].apply(lambda x: 1 if x=='yes' else 0)
data['activities']=data['activities'].apply(lambda x: 1 if x=='yes' else 0)
data['nursery']=data['nursery'].apply(lambda x: 1 if x=='yes' else 0)
data['higher']=data['higher'].apply(lambda x: 1 if x=='yes' else 0)
data['internet']=data['internet'].apply(lambda x: 1 if x=='yes' else 0)
data['romantic']=data['romantic'].apply(lambda x: 1 if x=='yes' else 0)
```

```python
data.head()
```

```python
Mjob=pd.get_dummies(data['Mjob'],dtype=int)
Fjob=pd.get_dummies(data['Fjob'],dtype=int)
reason=pd.get_dummies(data['reason'],dtype=int)
guardian=pd.get_dummies(data['guardian'],dtype=int)
```

```python
data=pd.concat([data.drop('Mjob',axis=1),Mjob],axis=1)
data=pd.concat([data.drop('Fjob',axis=1),Fjob],axis=1)
data=pd.concat([data.drop('reason',axis=1),reason],axis=1)
data=pd.concat([data.drop('guardian',axis=1),guardian],axis=1)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split

train_data,test_data=train_test_split(data,test_size=0.2)
```

```python
train_data
```

```python
test_data
```

```python
x_train=train_data.drop('G3',axis=1)
y_train=train_data['G3']

x_test=test_data.drop('G3',axis=1)
y_test=test_data['G3']
```

```python
x_train.head()
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestRegressor 

forest=RandomForestRegressor()

forest.fit(x_train,y_train)
```

```python
forest.score(x_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid={
    'n_estimators':[300,250,200],
    'max_features':[20,35,14],
    'max_depth':[2,10,20]
}

grid_search=GridSearchCV(forest,cv=15,param_grid=param_grid,scoring='neg_mean_squared_error',verbose=10,return_train_score=True)
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_estimator_.score(x_test,y_test)
```
